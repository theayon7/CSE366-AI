```python
import random

#Set the number
nq = 4  #(4x4)

#random chromosome
def random_chromosome(size=nq):
    return [random.randint(1, size) for _ in range(size)]

#Fitness function
def fitness(chromosome):
    score = 0
    for i in range(len(chromosome)):
        for j in range(i + 1, len(chromosome)):
            if chromosome[i] != chromosome[j] and abs(chromosome[i] - chromosome[j]) != abs(i - j):
                score += 1
    return score

#Selection
def selection(population):
    fitnesses = [fitness(chromo) for chromo in population]
    total_fitness = sum(fitnesses)
    if total_fitness == 0:
        return random.choice(population)
    pick = random.uniform(0, total_fitness)
    current = 0
    for chromo, fit in zip(population, fitnesses):
        current += fit
        if current > pick:
            return chromo

#Crossover
def crossover(parent1, parent2):
    point = random.randint(1, nq - 1)
    return parent1[:point] + parent2[point:]

#Mutation
def mutate(chromosome, mutation_rate=0.1):
    if random.random() < mutation_rate:
        index = random.randint(0, nq - 1)
        chromosome[index] = random.randint(1, nq)
    return chromosome

#printing the board
def print_board(chromosome):
    print("Chess Board:")
    for i in range(1, nq+1):
        row = ""
        for j in range(nq):
            if chromosome[j] == i:
                row += " Q "
            else:
                row += " . "
        print(row)
    print()

#Genetic Algorithm function
def genetic_algorithm(pop_size=100, generations=1000):
    population = [random_chromosome() for _ in range(pop_size)]
    max_fitness = (nq * (nq - 1)) // 2 

    for gen in range(generations):
        for chromo in population:
            if fitness(chromo) == max_fitness:
                print(f"✅ Solution found in generation {gen}: {chromo}")
                print_board(chromo)
                return chromo

        new_population = []
        for _ in range(pop_size):
            parent1 = selection(population)
            parent2 = selection(population)
            child = crossover(parent1, parent2)
            child = mutate(child)
            new_population.append(child)

        population = new_population

    print("No solution found.")
    return None

# Step 9: Run the algorithm
solution = genetic_algorithm()
```
